# Práctica de Python: If y Else

En esta práctica aprenderás a usar las estructuras condicionales **if** y **else** en Python.

Recuerda:
```python
if condición:
    # código si la condición es verdadera
else:
    # código si la condición es falsa
```

## Ejercicio 1
Pide al usuario que ingrese su edad y muestra:
- 'Eres mayor de edad' si tiene 18 años o más.
- 'Eres menor de edad' si tiene menos de 18.

Escribe tu código en la celda siguiente:

````python
# Escribe tu código aquí
# Pedir al usuario que ingrese su edad
edad = int(input("Ingresa tu edad: "))

# Verificar si es mayor o menor de edad
if edad >= 18:
    print("Eres mayor de edad")
else:
    print("Eres menor de edad")
````

## Ejercicio 2
Pide al usuario que ingrese un número. Si el número es **positivo** muestra 'Es positivo', si es **negativo** muestra 'Es negativo', si es **cero** muestra 'Es cero'.

Escribe tu código en la celda siguiente:

````python
# Escribe tu código aquí
# Pedir al usuario que ingrese un número
numero = float(input("Ingresa un número: "))

# Evaluar si es positivo, negativo o cero
if numero > 0:
    print("Es positivo")
elif numero < 0:
    print("Es negativo")
else:
    print("Es cero")
````

## Ejercicio 3
Pide al usuario que ingrese una contraseña. Si la contraseña es `'python123'` muestra 'Acceso permitido'. Si no, muestra 'Contraseña incorrecta'.

Escribe tu código en la celda siguiente:

````python
# Pedir al usuario que ingrese una contraseña
contraseña = input("Ingresa la contraseña: ")

# Verificar si la contraseña es correcta
if contraseña == "python123":
    print("Acceso permitido")
else:
    print("Contraseña incorrecta")
````

## Ejercicio 4
Pide al usuario que ingrese su calificación (0 a 100). Si la calificación es **mayor o igual a 60**, imprime 'Aprobado'. De lo contrario, imprime 'Reprobado'.

Escribe tu código en la celda siguiente:

````python
# Pedir al usuario que ingrese su calificación
calificacion = float(input("Ingresa tu calificación (0 a 100): "))

# Verificar si aprobó o reprobó
if calificacion >= 60:
    print("Aprobado")
else:
    print("Reprobado")
````

## Ejercicio 5
Pide al usuario que escriba un color. Si el color es `'rojo'`, `'azul'` o `'verde'`, muestra 'Es un color primario'. Si no, muestra 'No es un color primario'.

Escribe tu código en la celda siguiente:

````python
# Escribe tu código aquí
# Pedir al usuario que escriba un color
color = input("Escribe un color: ").lower()  # Convertir a minúsculas para evitar errores por mayúsculas

# Verificar si es un color primario
if color == "rojo" or color == "azul" or color == "verde":
    print("Es un color primario")
else:
    print("No es un color primario")
````
