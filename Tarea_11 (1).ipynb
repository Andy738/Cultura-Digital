## Parte 3: Condicionales en Python

### Ejercicio 1
Pide la hora (0-23). Si es menor que 12 muestra 'Es mañana', si no 'Es tarde o noche'.


```python
# Escribe tu código aquí
hora = int(input("Ingresa la hora (0-23): "))

if hora < 12:
    print("Es mañana")
else:
    print("Es tarde o noche")
```

### Ejercicio 2
Pide dos números y muestra cuál es mayor o si son iguales.


```python
# Escribe tu código aquí
num1 = float(input("Ingresa el primer número: "))
num2 = float(input("Ingresa el segundo número: "))

if num1 > num2:
    print("El primero es mayor")
elif num2 > num1:
    print("El segundo es mayor")
else:
    print("Son iguales")
```

### Ejercicio 3
Pide un número entero y muestra si es par o impar.


```python
# Escribe tu código aquí
numero = int(input("Ingresa un número entero: "))

if numero % 2 == 0:
    print("Es par")
else:
    print("Es impar")
```
