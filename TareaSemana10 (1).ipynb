### ***Ejercicio 1:***  
Pide dos números al usuario y muestre en pantalla el resultado de multiplicarlos.

````python
#Respuestas Ejercicio1
# Pedir dos números al usuario
num1 = float(input("Introduce el primer número: "))
num2 = float(input("Introduce el segundo número: "))

# Multiplicar los números
resultado = num1 * num2

# Mostrar el resultado
print("El resultado de la multiplicación es:", resultado)

````

### ***Ejercicio 2***:

### ***Ejercicio 3:***
Pide al usuario su nombre y apellido, y muestra el nombre completo.

### ***Ejercicio 4:***
Pide al usuario el precio de un producto y la cantidad que va a comprar, luego muestra el total a pagar.

````python
#Respuesta de ejercicio 4
# Pedir el precio y la cantidad al usuario
precio = float(input("Introduce el precio del producto: "))
cantidad = int(input("Introduce la cantidad que vas a comprar: "))

# Calcular el total
total = precio * cantidad

# Mostrar el total a pagar
print(f"El total a pagar es: {total}")
````

### ***Ejercicio 5:***
Pide al usuario una cantidad de minutos y muestra cuántos segundos equivalen.

````python
#Respuesta de ejercicio 5
# Pedir una cantidad de minutos al usuario
minutos = float(input("Introduce una cantidad de minutos: "))

# Calcular los segundos equivalentes
segundos = minutos * 60

# Mostrar el resultado
print(f"{minutos} minutos equivalen a {segundos} segundos.")
````
